Criar um sistema bancário com python

Objetivo geral: criar um sistema bancário com operações de sacar, depositar e visualizar texto.

```python
#Menu interativo para escolha das opções
menu = """

[1] Depositar
[2] Sacar
[3] Extrato
[4] Sair

=> """

#declaração das variáveis
saldo = 0
limite = 500
extrato = ""
numero_saque = 0
limite_saque = 3

while True:

  opcao = input(menu) #variavel que recebe as opções de menu

  if opcao == "1":
    valor = float(input("Informe o valor do depósito: "))

    if valor > 0:
      saldo += valor
      extrato += f"Depósito R$ {valor:.2f}\n"

    else:
      print("ERRO! Valor informado é inválido")

  elif opcao == "2":
      valor = float(input("Informe o valor para saque: "))

      saldo_excedeu = valor > saldo

      limite_excedeu = valor > limite

      saque_excedeu = numero_saque >= limite_saque

      if saldo_excedeu:
        print("ERRO! Você não possui saldo suficiente.")

      elif limite_excedeu:
        print("ERRO! Valor do saque é maior que limite disponível.")

      elif saque_excedeu:
        print("ERRO! Número de saques excedido.")

      elif valor > 0:
        saldo -= valor
        extrato += f"Saque: R$ {valor:.2f}\n"
        numero_saque += 1

  elif opcao == "3":
      print("\n================ EXTRATO ================")
      print("Não foram realizadas movimentações." if not extrato else extrato)
      print(f"\nSaldo: R$ {saldo:.2f}")
      print("==========================================")

  elif opcao == "4":
    break

  else:
      print("Operação inválida, por favor selecione novamente a operação desejada.")





```
